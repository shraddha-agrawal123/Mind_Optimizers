```python
import nltk
from nltk.chat.util import Chat, reflections
# Define patterns for the chatbot to recognize
patterns = [
    (r'hi|hello|hey', ['Hello!', 'Hi there!', 'How can I assist you today?']),
    (r'bye|goodbye', ['Goodbye!', 'Have a great day!']),
    (r'how are you?', ['I am just a chatbot, but thanks for asking!']),
    (r'(.*) order (.*)', ['Could you please provide your order number?']),
    (r'cancel order (.*)', ['To cancel your order, please contact our customer support team at support@example.com.']),
    (r'status (.*) order', ['To check the status of your order, please provide your order number.']),
    (r'order (.*) delayed', ['I apologize for the inconvenience. Please provide your order number so I can check the status.']),
    (r'product (.*) available', ['Please provide the name or ID of the product you are inquiring about.']),
   (r'(.*) out of stock', ["I'm sorry to hear that. We'll restock it soon!"]),
   (r'(.*) complaint (.*)', ["I'm sorry to hear that. Please email us at support@example.com with your complaint."]),
    (r'(.*) complaint (.*)', ['I\'m sorry to hear that. Please email us at support@example.com with your complaint.']),
    (r'(.*) return (.*)', ["For returns, please visit our Returns & Refunds page or contact our support team."]),
   (r'(.*) thank (.*)', ['You\'re welcome!', 'Happy to help!']),
     (r'hi|hello|hey', ['Hello!', 'Hi there!', 'How can I assist you today?']),
    (r'(.*) help (.*)', ['Sure, I can help you. What do you need assistance with?']),
    (r'(.*) your name?', ["I'm just a chatbot.", "I don't have a name."]),
    (r'(.*) order (.*)', ['For assistance with orders, please provide your order number or contact our support team.']),
    (r'(.*) product (.*)', ['To find a product, please visit our website or use the search function.']),
    (r'(.*) return (.*)', ["For returns, please visit our Returns & Refunds page or contact our support team."]),
    (r'(.*) complaint (.*)', ["I'm sorry to hear that. Please email us at support@example.com with your complaint."]),
    (r'(.*) out of stock', ["I'm sorry to hear that. We'll restock it soon!"]),
    (r'(.*) thank (.*)', ["You're welcome!", 'Happy to help!']),
    (r'quit', ['Goodbye!', 'See you later.']),
]
cb = Chat(patterns, reflections)
def chat():
    print("Welcome to our e-commerce customer support. How can I assist you today?")
    while True:''
        user_input = input("You: ")
        if user_input.lower() == 'quit':
            print("Chatbot: Goodbye!")
            break
        else:
            response = cb.respond(user_input)
            print("Chatbot:", response)
if __name__ == "__main__":
    nltk.download('punkt')
    nltk.download('averaged_perceptron_tagger')
    chat()
```
